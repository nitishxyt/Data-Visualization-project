```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data = data.dropna(axis=0)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
# Bar chart for top 10 movies by revenue
top_10_revenue = data.nlargest(10, 'Revenue (Millions)')
plt.figure(figsize=(10, 6))
plt.bar(top_10_revenue['Title'], top_10_revenue['Revenue (Millions)'], color='skyblue')
plt.title('Top 10 Movies by Revenue')
plt.xlabel('Movie Title')
plt.ylabel('Revenue (Millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
 # Split the 'Genre' column by commas and count the occurrences of each genre
genre_counts = data['Genre'].str.split(',').explode().value_counts()
plt.figure(figsize=(10, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Genre Distribution of Movies')
plt.tight_layout()
plt.show()
```

```python
# Calculate the average rating per year
average_rating_per_year = data.groupby('Year')['Rating'].mean()
# Line chart for average rating over years
plt.figure(figsize=(8, 6))
plt.plot(average_rating_per_year.index, average_rating_per_year.values, marker='o', color='green')
plt.title('Average Movie Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Rating vs. Revenue (Millions)
plt.figure(figsize=(10, 6))
plt.scatter(data['Rating'], data['Revenue (Millions)'], color='purple', alpha=0.5)
plt.title('Scatter Plot: Movie Rating vs. Revenue')
plt.xlabel('Rating')
plt.ylabel('Revenue (Millions)')
plt.grid()
plt.tight_layout()
plt.show()
```
